# Linux跨平台二进制函数识别

```python
import json
import csv
import numpy as np
import pandas as pd
```

```python
file_path = 'E:/Dataset/linux_binary'
```

```python
def read_files(file):
    with open(file) as f:
        for line in f.readlines():
            x = json.loads(line)
            yield x
```

```python
train_func = read_files(f"{file_path}/train.func.json")
test_func = read_files(f"{file_path}/test.func.update.json")
```

```python
dic = {
    'arch': [],
    'compiler': [], 
    'opti': []
}
i = 0
for func in train_func:
    for key in ['arch', 'compiler', 'opti']:
        if func[key] not in dic[key]:
            dic[key].append(func[key])
    i += 1
print(dic)
i
```

```python
max_num = 0
min_num = 1
for func in train_func:
    min_num = len(func['block']) if len(func['block']) < min_num else min_num
    max_num = len(func['block']) if len(func['block']) > max_num else max_num
print(min_num, max_num)
```

```python
dic = {
    'arch': [],
    'compiler': [], 
    'opti': []
}
i = 0
for func in test_func:
    for key in ['arch', 'compiler', 'opti']:
        if func[key] not in dic[key]:
            dic[key].append(func[key])
    i += 1
i
```

```python
max_num = 0
min_num = 1
for func in test_func:
    min_num = len(func['block']) if len(func['block']) < min_num else min_num
    max_num = len(func['block']) if len(func['block']) > max_num else max_num
print(min_num, max_num)
```

```python
max_num = 0
min_num = 1
with open(f"{file_path}/train.group.csv") as f:
    reader = csv.reader(f)
    for row in reader:
        min_num = len(row) if len(row) < min_num else min_num
        max_num = len(row) if len(row) > max_num else max_num
print(row)
print(reader.line_num, min_num, max_num)
```
